```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('/Users/akshay/Downloads/ecommerce/seeds/ecom_data.csv')
```

```python
new_df = df.loc[:1000].dropna(axis=1)
```

```python
new_df = new_df[['prices.amountMax','name', 'primaryCategories','imageURLs','categories','brand','weight','prices.merchant',]]
```

```python
combine_lambda = lambda x : f"This product is made by {x.brand}, it wieghs {x.weight}, it is sold by {x['prices.merchant']} Categories:{x.categories}" 

    
```

```python
new_df['description'] = (new_df.apply(combine_lambda,axis = 1))
```

```python
new_df.imageURLs = new_df.imageURLs.apply(lambda x: x.split(','))
```
